```python
images_train, cls_train, labels_train = cifar10.load_training_data()
images_test, cls_test, labels_test = cifar10.load_test_data()
```

```python
from sklearn import ensemble
from sklearn.metrics import classification_report,confusion_matrix
images_train= images_train.reshape(images_train.shape[0], -1)
images_test= images_test.reshape(images_test.shape[0], -1)
rf=ensemble.RandomForestClassifier()
rf.fit(images_train,cls_train)
y_predc=rf.predict(images_test)
#print(classification_report(images_test,cls_test))
#print(confusion_matrix(images_test,cls_test))
print(rf.score(images_test,cls_test))
y_predcc=y_predc.reshape(-1,1)
import numpy as np
import pandas as pd
ypredccc=np.savetxt('ypredccc.csv',y_predcc,delimiter=",")
cifar=pd.read_csv("ypredccc.csv")
print(cifar.shape)

print(cifar.shape)
print(cls_test.shape)
a=open("ypredccc.csv",'r')
print("the file ")
print(a.read())
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier()
images_test= images_test.reshape(images_test.shape[0], -1)
clf.fit(images_test,cls_test)
y_predclf=clf.predict(images_test)
print(clf.score(images_test,cls_test))
```

```python
def f(s):
  if(s==0):
    return "airplane"
  if(s==1):
    return "automobile"
  if(s==2):
    return "bird"
  if(s==3):
    return "cat"
  if(s==4):
    return "deer"
  if(s==5):
    return "dog"
  if(s==6):
    return "frog"
  if(s==7):
    return "horse"
  if(s==8):
    return "ship"
  if(s==9):
    return "truck"
cifar.columns=['class']
print(cifar.shape)
print(cifar.columns)

cifar["class1"]=cifar["class"].apply(f)
cifar.columns=['class','airplane']
del cifar['class']
print(cifar)

```
