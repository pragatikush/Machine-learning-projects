```python
def step_gradient(points,learn_rate,m):
    m_slope=np.zeros(5)
    M=len(points)
    for i in range(M):
        y=points[i,5]
        s=0
        for j in range(5):
           s=s+(m[j]*points[i,j]) 
        for k in range(5):
           m_slope[k]+=(-2/M)*(y-s)*points[i,k]
        new_m=m-learn_rate*m_slope
    return new_m
```

```python
def gd(points,learn_rate,num_itr):
    m=np.zeros(5)
    for i in range(num_itr):
        m=step_gradient(points,learn_rate,m)
       # print(i,"cost ",cost(points,m))
    return m
```
